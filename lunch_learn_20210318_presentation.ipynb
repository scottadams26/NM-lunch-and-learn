## Visualizations

The pandas library includes some built-in functions for data visualization. While an overview of all of pandas' visualization capabilities are beyond the scope of this session, more information can be found [here](https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html) at the pandas webpage. I encourage you to try out some other visualization techniques on your own. For now, let's create a histogram of the median housing value. Note that we can specify the column we want to plot, here, by placing the string of the column name in brackets immediately following the name of the data frame. 

```python
df["median_house_value"].plot.hist()
```
