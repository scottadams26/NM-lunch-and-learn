**NOTE** The `as pd` in `import pandas as pd` is an **alias**. For example, `read_csv` is the pandas function to read a comma separated value (CSV) file. To use this function we would need to write `pd.read_csv`. Had we imported pandas without an alias as `import pandas`, then we would need to write `pandas.read_csv`. Thus, the alias just saves us some typing.

## Data Exploration

We read our data in as a **data frame** stored in the variable `df`. A data frame is a tabular data storage object comprised of columns and rows, much like an Excel spreadsheet. However, while we can simply glance at the data in Excel by opening the data file in Excel, with Python we need to write code to explore the data. For example, to see the first five rows of data, we can use the `head` function, which gets applied directly to the data frame `df`.

```python
df.head()
```

A few notes, here. First, look at the row numbers on the left of the output. The row numbers range from 0-4, not 1-5. Why? Python is **zero-indexed**, meaning that the counting of elements in an object starts at `0`, not `1`. R, by contrast, is indexed at one. 

Second, five is the default number of rows returned by `head`, but you can specify as many rows as you want by passing the desired number of rows as an argument in the parentheses. Try returning the first 10 rows in the cell below.

What if we wanted to see the last 5 rows of data? Use the `tail` function instead of `head`. Try it yourself.

Alternatively, we could also use `loc` to return the first five row. `loc` has the following structure.

image.png

In the cell below, the `0:4` means that we want to return rows 0 (which is the first row in the data frame) to 4 (which is actually the fifth row), including row 4. Note that there are no columns specified in the code below, which means that all the columns in the data frame will be returned.


```python
df.loc[0:4, :]
```

While a discussion of the nuances of `loc` and it's use of square brackets is beyond the scope of this session, here is an [online post] (https://stackoverflow.com/questions/46176656/why-how-does-pandas-use-square-brackets-with-loc-and-iloc) that provides some more detail.

`loc` is particularly useful when we want to specific conditions on both rows and columns. For instance, if we wanted to return the first five rows and specific columns, then we would need to provide the write the column names as strings in a **list** (which, without getting into too much detail, is a collection of data values within a set of square brackets).

```python
df.loc[0:4, ['housing_median_age', "median_house_value"] ]
```
